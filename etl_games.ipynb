# ETL (games)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.max_columns = None
games = pd.read_csv('games.csv')
games_a = pd.read_csv('games_a.csv')
```

```python
g = pd.DataFrame(games)
ga =  pd.DataFrame(games_a)
```

```python
display(games.head(5))
```

```python
display(games_a.head(5))
```

## Renaming the home_club_id to club_id, and away_club_id to adversary_club_id, 

```python
# renaming the "home_club_id" columns in games df 

g.rename(columns={'home_club_id': 'club_id'}, inplace = True)
g.rename(columns={'away_club_id': 'adversary_club_id'}, inplace = True)


# renaming the "home_club_id" columns in games_a df

ga.rename(columns={'home_club_id': 'adversary_club_id'}, inplace = True)
ga.rename(columns={'away_club_id': 'club_id'}, inplace = True)

# renaming the "home_club_goals" columns in games df

g.rename(columns={'home_club_goals': 'club_goals'}, inplace = True)
g.rename(columns={'away_club_goals': 'adversary_club_goals'}, inplace = True)


# renaming the "home_club_goals" columns in games_a df

ga.rename(columns={'home_club_goals': 'adversary_club_goals'}, inplace = True)
ga.rename(columns={'away_club_goals': 'club_goals'}, inplace = True)

# renaming the "home_club_position" columns in games df

g.rename(columns={'home_club_position': 'club_position'}, inplace = True)
g.rename(columns={'away_club_position': 'adversary_club_position'}, inplace = True)


# renaming the "home_club_position" columns in games_a df
ga.rename(columns={'home_club_position': 'adversary_club_position'}, inplace = True)
ga.rename(columns={'away_club_position': 'club_position'}, inplace = True)
```

```python
display(g.head(5))
```

```python
display(ga.head(5))
```

```python
ga = ga[['game_id', 'competition_code',	'season',	'round',	'date',	'club_id', 'adversary_club_id', 'club_goals',	'adversary_club_goals', 'club_position',	'adversary_club_position',	'stadium',	'attendance',	'referee',	'url']]

# printing the altered DataFrame

print('After altering Name and Maths Score')
display(ga.head(5))
```

```python
## Adding the columns Home/Away in the DF's
g = g.assign(h_a='home')
display(g.head(5))
```

```python
ga = ga.assign(h_a='away')
display(ga.head(5))
print("ga.shape: ({} x {})".format(ga.shape[0], ga.shape[1]))
```

## Merging the DF's

```python
g_merged = pd.concat(objs=[g, ga], axis=0).reset_index(drop=True)
print("g_merged.shape: ({} x {})".format(g_merged.shape[0], g_merged.shape[1]))
```

```python
g_merged = g_merged.assign(result=0)
display(g_merged.head(5))
```

```python
display(g_merged.dtypes)
display(g_merged.head(10))
```

## Adding the Results columns

```python
g_merged['club_goals'] = g_merged['club_goals'].astype(int)
g_merged['adversary_club_goals'] = g_merged['adversary_club_goals'].astype(int)
```

```python
g_merged['result'] = np.where(g_merged['club_goals'] > g_merged['adversary_club_goals'],1,np.where(g_merged['club_goals']  == g_merged['adversary_club_goals'],0,np.where(g_merged['club_goals'] < g_merged['adversary_club_goals'],-1,0)))
display(g_merged.head(10))
```

```python
g_merged.loc[g_merged.result == 1,'result_n']='win'
g_merged.loc[g_merged.result == 0,'result_n']='draw'
g_merged.loc[g_merged.result == -1,'result_n']='loss'
display(g_merged.head(10))
```

## Exporting the DF's as csv

```python
g_merged.to_csv('games1.csv')
```
